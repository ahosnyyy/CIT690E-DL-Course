# Nesterov Accelerated SGD

The SGD or Stochastic Gradient Optimizer is an optimizer in which the weights are updated for each training sample or a small subset of data.

### Syntax

The following shows the syntax of the Nesterov Accelerated SGD optimizer in PyTorch.

```python
torch.optim.SGD(params, lr=<required parameter>, momentum=gamma, dampening=0, weight_decay=0, nesterov=False)
```

### Parameters

* <b>params</b> (iterable) — These are the parameters that help in the optimization.
* <b>lr</b> (float) — This parameter is the learning rate
* <b>momentum</b> (float, optional) — Here we pass the momentum factor `default=0`
* <b>weight_decay</b> (float, optional) — This argument is containing the weight decay
* <b>dampening</b> (float, optional) — To dampen the momentum, we use this parameter
* <b>nesterov</b> (bool, optional) – enables Nesterov momentum `default: False`

# 2-D Function Optimization

````python
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
from IPython.display import HTML
````

# Optimization of function 
$ f(x) = x^{2} $

````python
def func2d(x):
    return x**2
````

````python
xmin,xmax,xstep = -10.5,10.5,0.2
x_points = np.arange(xmin,xmax,xstep)
f_x_2d = []
for i in range(0,len(x_points)):
    f_x_2d.append(func2d(x_points[i]))
````

# Pick a starting Point & Define minima of the function

````python
#x0 = random.randint(int(xmin),int(xmax))
x0 = -9
minima_x = 0
print('X0 : %d'%(x0))
print('Minima : %.2f'%(func2d(minima_x)))
````

# Plot Function Surface

````python
plt.plot(x_points,f_x_2d)
plt.xlabel('x')
plt.plot(x0,func2d(x0),'r*',markersize=10,label='Starting Point')
plt.plot(minima_x,func2d(minima_x),'*',markersize=10,label='Minima')
plt.legend()
plt.ylabel('F(x)')
plt.show()
````

# Build Computation Graph for Automatic Differentiation using PyTorch

````python
import torch
from IPython import display
import time

epochs = 200
learning_rate = 0.01
path_x, path_fx, vel_track = [], [], []
gamma = 0.9

# make sure you set initial point correctly
x = torch.tensor(x0,dtype=torch.float32,requires_grad=True)
optimizer = torch.optim.SGD([x],lr=learning_rate,momentum=gamma,nesterov=True)

for epoch in range(0,epochs):
    
    # loss
    loss = (x)**2

    # calculate gradients
    loss.backward()
    
    optimizer.step()
    
    with torch.no_grad():
        x_numpy = x.detach().numpy()
        path_x.append(np.copy(x_numpy))
        path_fx.append(func2d(np.copy(x_numpy)))
    
    x.grad.data.zero_()
````

````python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

def animate(i):
    ax.clear()
    ax.plot(x_points,f_x_2d,label='Surface of $x^{2}$, Step Size : '+(str(learning_rate)),color='red')
    ax.plot(minima_x,func2d(minima_x),'*',markersize=10,color='purple', zorder=9)
    ax.plot(path_x[0:i],path_fx[0:i],color='black',linewidth=4, zorder=10)
    ax.plot(path_x[i],path_fx[i],'o',color='purple', zorder=10)
    ax.legend()
    ax.set_xlabel('$x$')
    ax.set_ylabel('$f(x,y)$')
    ax.set_title('Nesterov SGD')

ani = animation.FuncAnimation(fig,animate,frames=len(path_x),interval=60,repeat=False)
````

````python
HTML(ani.to_html5_video())
````

````python
ani.save('convex_nesterov.gif',writer='pillow')
````

# Optimization of function
$ f(x) = log( 1 + |x| ) ^ {2 + sin(x)} $

````python
def func2d(x):
    return math.log(1 + math.fabs(x)**(2+math.sin(x)))
````

````python
xmin,xmax,xstep = -10.5,10.5,0.2
x_points = np.arange(xmin,xmax,xstep)
f_x_2d = []
for i in range(0,len(x_points)):
    f_x_2d.append(func2d(x_points[i]))
````

# Pick a starting Point & Define minima of the function

````python
x0 = random.randint(int(xmin),int(xmax))
x0 = 7
minima_x = 0
print('X0 : %d'%(x0))
print('Minima : %.2f'%(func2d(minima_x)))
````

# Ploting surface of the function

````python
plt.plot(x_points,f_x_2d)
plt.xlabel('x')
plt.plot(x0,func2d(x0),'r*',markersize=10,label='Starting Point')
plt.plot(minima_x,func2d(minima_x),'*',markersize=10,label='Minima')
plt.legend()
plt.ylabel('F(x)')
plt.show()
````

````python
import torch
from IPython import display
import time

epochs = 300
learning_rate = 0.01
path_x, path_fx, vel_track = [], [], []
gamma = 0.9

# make sure you set initial point correctly
x = torch.tensor(x0,dtype=torch.float32,requires_grad=True)
optimizer = torch.optim.SGD([x],lr=learning_rate,momentum=gamma,nesterov=True)

for epoch in range(0,epochs):
    
    # loss
    loss = torch.log(1 + (torch.abs(x)**(2 + torch.sin(x))))
    
    # calculate gradients
    loss.backward()
    
    # Update params
    optimizer.step()
    
    with torch.no_grad():
        x_numpy = x.detach().numpy()
        path_x.append(np.copy(x_numpy))
        path_fx.append(func2d(np.copy(x_numpy)))
    
    # make gradients zeros
    x.grad.data.zero_()
````

````python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

def animate(i):
    ax.clear()
    ax.plot(x_points,f_x_2d,label='Surface of $x^{2}$, Step Size : '+(str(learning_rate)),color='red')
    ax.plot(minima_x,func2d(minima_x),'r*',markersize=10,color='purple')
    ax.plot(path_x[0:i],path_fx[0:i],color='black',linewidth=4)
    ax.plot(path_x[i],path_fx[i],'o',color='purple')
    ax.legend()
    ax.set_xlabel('$x$')
    ax.set_ylabel('$f(x,y)$')
    ax.set_title('Nesterov SGD')

ani = animation.FuncAnimation(fig,animate,frames=len(path_x),interval=60,repeat=False)
````

# 3D function optimization
$ f(x) = x^2 + y^2 $

````python
f  = lambda x, y: (x**2) - (y**2)
````

````python
xmin,xmax,xstep = -4.5,4.5,0.2
ymin,ymax,ystep = -4.5,4.5,0.2
````

````python
x,y = np.meshgrid(np.arange(xmin,xmax,xstep),np.arange(ymin,ymax,ystep))
z = f(x,y)
````

````python
from mpl_toolkits.mplot3d import Axes3D
````

# Pick a starting Point & Define minima of the function

````python
x0 = -4
y0 = 0
print('X0 : %.2f, Y0 : %.2f '%(x0,y0))
````

# Plot function surface

````python
fig = plt.figure(figsize=(15,10))
ax = plt.axes(projection='3d')
ax.plot_surface(x,y,z,cmap='viridis')
ax.plot([x0],[y0],[f(x0,y0)],'o',color='purple', zorder=10)
ax.set_title('Surface of $f(x,y) = x^{2} - y^{2}$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
````

````python
import torch
from IPython import display
import time

epochs = 180
learning_rate = 0.01

path_x, path_y, path_fxy = [], [], []
gamma = 0.9

# make sure you set initial point correctly
x_torch = torch.tensor(x0,dtype=torch.float32,requires_grad=True)
y_torch = torch.tensor(y0,dtype=torch.float32,requires_grad=True)

optimizer = torch.optim.SGD([x_torch,y_torch],lr=learning_rate,momentum=gamma,nesterov=True)

for epoch in range(0,epochs):
    
    # objective function
    loss = (x_torch**2) - (y_torch**2)
    
    # backward
    loss.backward()
    
    # Update params
    optimizer.step()
    
    with torch.no_grad():
        x_numpy = x_torch.detach().numpy()
        y_numpy = y_torch.detach().numpy()
        
        path_x.append(np.copy(x_numpy))
        path_y.append(np.copy(y_numpy))
        path_fxy.append(np.copy(f(x_numpy,y_numpy)))
    
    # make gradients zero
    x_torch.grad.data.zero_()
    y_torch.grad.data.zero_()
````

````python
fig = plt.figure(figsize=(12,8))
ax = plt.axes(projection='3d')
ax.set_title('Nesterov SGD')
def animate(i):
    
    ax.clear()
    ax.plot_surface(x,y,z,cmap='viridis')
    ax.plot([x0],[y0],[f(x0,y0)],'o', color='purple', zorder=10)
    ax.plot(path_x[0:i],path_y[0:i],path_fxy[0:i],linewidth=2,color='black', zorder=10)
    ax.set_title('Surface of $f(x,y) = x^{2} - y^{2}$')
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$f(x,y)$')
    ax.set_label('Nesterov SGD')

ani = animation.FuncAnimation(fig,animate,frames=len(path_x),interval=60,repeat=True)
````

````python
HTML(ani.to_html5_video())
````

````python
ani.save('3d_surface_nesterov.gif',writer='pillow')
````
